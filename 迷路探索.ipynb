- `height`：迷路の縦（y方向）の大きさ
- `width`（横）：迷路の横（x方向）の大きさ

迷路の外側に通路（余白）あり。

迷路の大きさが5×7の場合、0行目、6行目、0列目、8列目は余白になる：
```
  0 1 2 3 4 5 6 7 8
0
1   # # # # # # #
2   S   #       #
3   #   #   #   #
4   #       #   G
5   # # # # # # #
6
```

````python
import random

# 辞書機能で迷路部品を作る
# 道 壁 スタート ゴール 通過した道
buhin = {0: " ", 1: "#", 2: "S", 3: "G", 4: "o"}

# 迷路の幅、高さ
height = int(input("迷路の縦?（必ず奇数）"))
width = int(input("迷路の横?（必ず奇数）"))
print()

# 迷路のデータ格納配列作成
maze = [[0] * (width + 2) for _ in range(height + 2)]

# 迷路表示
def maze_display():
    for y in range(1, height + 1):
        for x in range(1, width + 1):
            print(buhin[maze[y][x]] + " ", end = "")
        print()
    print()

# 上下左右に穴を伸ばす
def ana_check(x, y):
    directions = [(0, 2), (0, -2), (2, 0), (-2, 0)]
    random.shuffle(directions)
    for dx, dy in directions:
        nx, ny = x + dx, y + dy
        if maze[ny][nx] == 1:
            maze[ny][nx] = maze[y + dy // 2][x + dx // 2] = 0
            ana_check(nx, ny)


# 迷路を作成する（穴掘り法）
def maze_make():
    # 壁で初期化
    for y in range(1, height + 1):
        for x in range(1, width + 1):
            maze[y][x] = 1
    # 適当なセルに穴を掘る
    x = random.randrange(2, width + 1, 2)
    y = random.randrange(2, height + 1, 2)
    maze[y][x] = 0
    ana_check(x, y)
    # 余白を壁で埋める（解く際に迷路の外に進むのを防ぐ）
    for x in range(0, width + 2):
        maze[0][x] = 1
        maze[height + 1][x] = 1
    for y in range(0, height + 2):
        maze[y][0] = 1
        maze[y][width + 1] = 1
    # スタート、ゴールの設定
    maze[2][1] = 2
    maze[height - 1][width] = 3

maze_make()
maze_display()

# 探索済みのマスを記録するリスト
visited = []

# 迷路を解く（深さ優先探索）
def maze_solve():
    route = route_check(1, 2, [])
    for x, y in route:
        maze[y][x] = 4

def route_check(x, y, visited):
    # ゴールに着いたら、探索終了
    if maze[y][x] == 3:
        return [(x, y)]
    # 今いる場所を壁にする
    for nx, ny in [(x, y + 1), (x, y - 1), (x + 1, y), (x - 1, y)]:
        # 進む方向が壁か？
        if maze[ny][nx] == 1:
            continue
        # 進む方向が探索済みか？
        if (nx, ny) in visited:
            continue
        visited.append((nx, ny))
        path = route_check(nx, ny, visited)
        # 将来の探索でゴールに着いた時
        if path:
            return [(x, y)] + path
    # 行き止まりに着いた時
    return None

maze_solve()
maze_display()
````
